```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Just replace '|' with space — TF-IDF handles the rest
movie_df['genres'] = movie_df['genres'].str.replace('|', ' ')

# Vectorize
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(movie_df['genres'])

# Similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
genre_mood = {
    'Action': 'Excitement', 'Adventure': 'Excitement',
    'Animation': 'Joy', 'Children': 'Joy',
    'Comedy': 'Happy', 'Romance': 'Romantic',
    'Drama': 'Emotional', 'Horror': 'Thrilled',
    'Thriller': 'Thrilled', 'Mystery': 'Curious',
    'Crime': 'Curious', 'Sci-Fi': 'Wonder',
    'Fantasy': 'Wonder', 'Documentary': 'Inspired',
    'War': 'Reflective', 'History': 'Reflective',
    'Musical': 'Cheerful', 'Music': 'Cheerful',
    'Western': 'Adventurous', 'Film-Noir': 'Melancholic',
    'IMAX': 'Wonder',
}

def assign_mood(genres_str):
    if pd.isnull(genres_str):
        return 'Neutral'
    for genre in genres_str.split(' '):  # already replaced | with space
        if genre in genre_mood:
            return genre_mood[genre]
    return 'Neutral'

movie_df['mood'] = movie_df['genres'].apply(assign_mood)
movie_df[['title', 'genres', 'mood']].sample(10)
```

```python
movie_df = movie_df.reset_index(drop=True)
movie_df.info()
```

```python
def recommend_movie(mood, genre, year_from=1980, n = 5):
    filtered = movie_df[
        (movie_df['mood'] == mood) &
        (movie_df['genres'].str.contains(genre, case=False)) &
        (movie_df['year'] >= year_from)
    ]
    return filtered.sort_values('imdb_rating', ascending=False)\
                   [['title', 'year', 'genres', 'imdb_rating', 'mood']]\
                   .head(n)

# Test it
recommend_movie('Excitement', 'Action')
```
